```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_spss('COVID-19 niños - profesores 04-18-23 - recode coding errors.sav')
```

```python
pd.set_option('display.max_columns', None)
```

```python
df= df.iloc[:,124:]
```

```python
df
```

```python
df = pd.concat([df, pd.get_dummies(df['ManVsOther']).rename(columns = {1: 'M', 2: 'F'})], axis = 1)
df.drop(['ManVsOther'], axis=1, inplace = True)
df.drop(['Country'], axis=1, inplace = True)
df
```

```python
df = pd.concat([df, pd.get_dummies(df['PartneredVsNot']).rename(columns = {1: 'M', 2: 'F'})], axis = 1)
df.drop(['PartneredVsNot'], axis=1, inplace = True)
df = pd.concat([df, pd.get_dummies(df['Education']).rename(columns = {1: 'M', 2: 'F'})], axis = 1)
df.drop(['Education'], axis=1, inplace = True)
df = pd.concat([df, pd.get_dummies(df['TeachAge0_5']).rename(columns = {1: 'M', 2: 'F'})], axis = 1)
df.drop(['TeachAge0_5'], axis=1, inplace = True)
df = pd.concat([df, pd.get_dummies(df['TeachAge6_11']).rename(columns = {1: 'M', 2: 'F'})], axis = 1)
df.drop(['TeachAge6_11'], axis=1, inplace = True)  
df = pd.concat([df, pd.get_dummies(df['TeachAge12_16']).rename(columns = {1: 'M', 2: 'F'})], axis = 1)
df.drop(['TeachAge12_16'], axis=1, inplace = True)   
df = pd.concat([df, pd.get_dummies(df['TeachAge17_18']).rename(columns = {1: 'M', 2: 'F'})], axis = 1)
df.drop(['TeachAge17_18'], axis=1, inplace = True)
df.drop([0.0], axis=1, inplace = True)  
df.drop(['PHQ1'], axis=1, inplace = True)  
df.drop(['PHQ2'], axis=1, inplace = True)
df.drop(['GAD1'], axis=1, inplace = True)
df.drop(['GAD2'], axis=1, inplace = True)
df.drop(['Q62_1Recode'], axis=1, inplace = True)
df.drop(['Q62_2Recode'], axis=1, inplace = True)
df.drop(['Q62_3Recode'], axis=1, inplace = True)
```
